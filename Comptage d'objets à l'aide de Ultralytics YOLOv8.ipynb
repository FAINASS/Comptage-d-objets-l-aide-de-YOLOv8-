# **Détection d'objets**

La détection d'objets est une tâche qui consiste à identifier l'emplacement et la classe des objets dans une image ou un flux vidéo.

```python
# Importation des bibliothèques nécessaires
from ultralytics import YOLO
from ultralytics.solutions import object_counter
import cv2

# Chargement du modèle YOLO
model = YOLO("yolov8n.pt")

# Ouverture du fichier vidéo
cap = cv2.VideoCapture("/content/Video.mp4")

# Vérification que le fichier vidéo a bien été ouvert
assert cap.isOpened(), "Error reading video file"

# Récupération des propriétés de la vidéo (largeur, hauteur, fps)
w, h, fps = (int(cap.get(x)) for x in (cv2.CAP_PROP_FRAME_WIDTH, cv2.CAP_PROP_FRAME_HEIGHT, cv2.CAP_PROP_FPS))

# Définition des points de la ligne de comptage et des classes à compter
line_points = [(1100, 1), (1100, 2800)]
classes_to_count = [0, 39]

# Initialisation de l'écriture vidéo
video_writer = cv2.VideoWriter("object_counting_output.avi",
                       cv2.VideoWriter_fourcc(*'mp4v'),
                       fps,
                       (w, h))

# Initialisation du compteur d'objets
counter = object_counter.ObjectCounter()
counter.set_args(view_img=True,
                 reg_pts=line_points,
                 classes_names=model.names,
                 draw_tracks=False)

# Boucle de lecture et de traitement de chaque image de la vidéo
while cap.isOpened():
    success, im0 = cap.read()
    if not success:
        print("Video frame is empty or video processing has been successfully completed.")
        break
    # Suivi des objets dans l'image
    tracks = model.track(im0, persist=True, show=False,
                         classes=classes_to_count)

    im0 = counter.start_counting(im0, tracks)
    video_writer.write(im0)

cap.release()
video_writer.release()
cv2.destroyAllWindows()
```
